**Under Sampling**

Build a sample dataset containing similar distribution of genuin and fraud transactons

Number of genuine  transactions ---

```python
genuine_sample = genuine.sample(n=492 )
```

```python
print(genuine_sample)
```

**CONCATENATING THE DATAFRAMES**

```python
new_dataset = pd.concat([genuine_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
  new_dataset.groupby('Class').mean()
```

Splititing the data into Features(y) and Targets((x)

```python
x = new_dataset.drop(columns='Class', axis=1)
y = new_dataset['Class']
x,y
```

splitting the dataset into the training and testing the data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,  random_state=2)
x_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
  model.fit(x_train, y_train)
```

MODEL EVALUATION


evaluating the accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
  x_train_prediction = model.predict(x_train)
  training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
training_data_accuracy*100
```

```python
  x_test_prediction = model.predict(x_test)
  test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
test_data_accuracy*100
```
